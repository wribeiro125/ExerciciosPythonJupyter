```python
# DESAFIO: Loteria
# Faça um programa em Python que simule uma loteria, gerando 6 números aleatórios, de forma que,
# os números não sejam repetidos e estejam entre 0 e 100,
# mostrando-os na tela e a quantidade de acertos do usuário

from random import sample
aposta = []
numero_de_acertos = 0
loteria = sample(range(100), 6)

for i in range(6):
    while(True):
        numero = int(input(f'Digite o {i+1}º número: '))
        if numero > 100 or numero < 0:
            print('Número não permitido')
        else:
            if numero in aposta:
                print('Número já digitado.')
            else:
                aposta.append(numero)
                break
            
for numero in loteria:
    if numero in aposta:
        numero_de_acertos += 1
aposta.sort()
loteria.sort()
print('')
print(f'Números sorteados: {loteria}')
print(f'Sua aposta: {aposta}')
print(f'Total de acertos: {numero_de_acertos}')
```
